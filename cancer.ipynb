```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dt=pd.read_csv("data.csv")
dt
```

```python
#EDA
dt.isnull().sum()
```

```python
dt.drop(['id','Unnamed: 32'],axis=1,inplace=True)
```

```python
dt
```

```python
dt.isnull().sum()
```

```python
dt.info()
```

```python
dt['diagnosis'] = dt['diagnosis'].map({'M':1,'B':0})
dt.head()
```

```python
X = dt[['texture_mean','area_mean','concavity_mean','area_se','concavity_se','fractal_dimension_se','smoothness_worst','concavity_worst', 'symmetry_worst','fractal_dimension_worst']]
```

```python
# Putting response variable to y
y = dt['diagnosis']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
sc=StandardScaler()
s_train=sc.fit_transform(x_train,y_train)
s_test=sc.transform(x_test)
```

```python
import warnings
from sklearn.svm import SVC
warnings.filterwarnings('ignore')
logmodel = SVC()
logmodel.fit(s_train,y_train)
```

```python
y_pred=logmodel.predict(s_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
#Finally save the model and Transformed data into Pickles
import pickle
```

```python
pickle.dump(logmodel,open('regmodel.pkl','wb'))
```

```python
#In the same way do for Feature Scaling part
pickle.dump(sc,open('scaling.pkl','wb'))
```
